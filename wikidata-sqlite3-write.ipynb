```python
import os
import gc
import ctypes
libc = ctypes.CDLL("libc.so.6")

import numpy as np
import pandas as pd
from tqdm.auto import tqdm

import sqlite3
from sqlalchemy import create_engine
```

```python
# 读取 wiki 文章数据

wikidata = []
list_dir = os.listdir('./wiki_data')
list_dir = sorted(list_dir)
for path in list_dir:
    if 'wiki_2023_index' in path or 'debert' in path or 'my' in path or 'ipynb' in path:
        continue
    print(path)
    x = pd.read_parquet(os.path.join('./wiki_data', path), 
                        columns=['id', 'text'])
    x['id'] = x['id'].astype(int)
    wikidata.append(x)
    del x
    _ = gc.collect()
    libc.malloc_trim(0)
wikidata = pd.concat(wikidata, axis=0).reset_index(drop=True)
wikidata['page_id'] = wikidata['id'].astype(int)
wikidata['id'] = wikidata.index
wikidata.set_index('id', inplace=True)

display(wikidata)
```

```python
df = wikidata[['text']]
df
```

```python
disk_engine = create_engine('sqlite:///wiki_text.db')
df.to_sql('wiki_text', disk_engine, if_exists='append')
```

```python
df = wikidata[['page_id']]
df
```

```python
disk_engine = create_engine('sqlite:///wiki_index.db')
df.to_sql('wiki_index', disk_engine, if_exists='append')
```
